```python
import numpy as np

# Hawk-Dove Game

V = 0
C = 0

def case1():
    global V, C
    V = 1
    C = 2
    
def case2():
    global V, C
    V = 1
    C = 5
    
def case3():
    global V,C
    V = 2
    C = 100
    
# Initial Population
population = {
    "homo_economicus": 1000,
    "homo_milliensis": 1000,
    "homo_kantiensis": 1000,
    "homo_smithiensis": 1000
}

# Set Population
def reset_population(E, M, K, S):
    population["homo_economicus"] = E
    population["homo_milliensis"] = M
    population["homo_kantiensis"] = K
    population["homo_smithiensis"] = S

# Individual Round
def simulate_round(population):
    total_population = sum(population.values())
    fitness = {key: 0 for key in population}
    for type1 in population:
        for type2 in population:
            # θ x θ Interactions
            interactions = population[type1] * population[type2]

            # Strategies
            
            strategy1 = 0
            strategy2 = 0
            u1 = 0

            ME1 = (V/C)**2 * ((V-C)/2) + ((C-V)/C)*(V/C)*0 + (V/C)*((C-V)/C)*V + ((C-V)/C)**2 * (V/2)
            ME2 = (V/2) * ((C-V)/C)
            ME3 = (V/C) * V + ((C-V)/C) * (V/2)
            
            if type1 == "homo_economicus":
                    if type2 == "homo_economicus":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * V + (1/3) * 0 + (1/3) * ME1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * V + (1/3) * 0 + (1/3) * ME2
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 1
                        strategy2 += 0
            elif type1 == "homo_milliensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * V + (1/3) * 0 + (1/3) * ME3
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * V + (1/3) * 0 + (1/3) * (V/2)
                    if type2 == "homo_kantiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * V + (1/2) * (V/2)
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * V + (1/2) * (V/2)
            elif type1 == "homo_kantiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * 0 + (1/2) * (V/2)
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0
                        strategy2 += 0
            if type1 == "homo_smithiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * 0 + (1/2) * (V/2)
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0
                        strategy2 += 0
       
            # Payoffs
            u1 += V/2 if strategy1 == 0 and strategy2 == 0 else 0
            u1 += 0 if strategy1 == 0 and strategy2 == 1 else 0 
            u1 += V if strategy1 == 1 and strategy2 == 0 else 0
            u1 += (V-C)/2 if strategy1 == 1 and strategy2 == 1 else 0
            
            # Update Fitness
            fitness[type1] += u1 * interactions

    # Update Population
    total_fitness = sum(fitness.values())
    for type in population:
        population[type] = int((fitness[type] / total_fitness) * total_population)
    return population

# Iteration
def simulate(A):
    global population
    num_generations = A
    for _ in range(num_generations):
        population = simulate_round(population)
    return population

def test():
    reset_population(1000,1000,0,0)
    simulate(1000)
    print("e x m = ", population)

    reset_population(1000,0,1000,0)
    simulate(1000)
    print("e x k = ", population)

    reset_population(1000,0,0,1000)
    simulate(1000)
    print("e x s = ", population)

    reset_population(0,1000,1000,0)
    simulate(1000)
    print("m x k = ", population)

    reset_population(0,1000,0,1000)
    simulate(1000)
    print("m x s = ", population)

    reset_population(0,0,1000,1000)
    simulate(1000)
    print("k x s = ", population)

    reset_population(1000,1000,1000,0)
    simulate(1000)
    print("e x m x k = ", population)

    reset_population(1000,1000,0,1000)
    simulate(1000)
    print("e x m x s = ", population)

    reset_population(1000,0,1000,1000)
    simulate(1000)
    print("e x k x s = ", population)

    reset_population(1000,1000,1000,0)
    simulate(1000)
    print("m x k x s = ", population)

    reset_population(1000,1000,1000,1000)
    simulate(1000)
    print("e x m x k x s", population)

def edge_test(a,b):
    reset_population(a,b,0,0)
    simulate(1000)
    print("e x m = ", population)

    reset_population(a,0,b,0)
    simulate(1000)
    print("e x k = ", population)

    reset_population(a,0,0,b)
    simulate(1000)
    print("e x s = ", population)

    reset_population(b,a,0,0)
    simulate(1000)
    print("m x e = ", population)
    
    reset_population(0,a,b,0)
    simulate(1000)
    print("m x k = ", population)

    reset_population(0,a,0,b)
    simulate(1000)
    print("m x s = ", population)
    
    reset_population(b,0,a,0)
    simulate(1000)
    print("k x e = ", population)
    
    reset_population(0,b,a,0)
    simulate(1000)
    print("k x m = ", population)

    reset_population(0,0,a,b)
    simulate(1000)
    print("k x s = ", population)

    reset_population(b,0,0,a)
    simulate(1000)
    print("s x e = ", population)
    
    reset_population(0,b,0,a)
    simulate(1000)
    print("s x m = ", population)

    reset_population(0,0,b,a)
    simulate(1000)
    print("s x k = ", population)
    
def test_all():
    case1()
    print("Case 1: C = 2, V = 1")
    test()

    
    case2()
    print("Case 2: C = 5, V = 1")
    test()


    case3()
    print("Case 3: C = 100, V = 2")
    test()
    
    
def test_skew(a,b):
    case1()
    print("Case 1: C = 2, V = 1")
    edge_test(a,b)

    case2()
    print("Case 2: C = 5, V = 1")
    edge_test(a,b)

    case3()
    print("Case 3: C = 100, V = 2")
    edge_test(a,b)

test_all()
```
