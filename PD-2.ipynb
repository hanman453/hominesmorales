```python
import numpy as np

# Prisoner's Dilemma

def lay(t, r, p, s):
    global T, R, P, S
    T = t
    R = r
    P = p
    S = s

def case1():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 6   # Reward for Mutual Cooperation
    P = 4   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation

def case2():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 8   # Reward for Mutual Cooperation
    P = 6   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case3():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 4   # Reward for Mutual Cooperation
    P = 2   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case4():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 5   # Reward for Mutual Cooperation
    P = 3   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case5():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 7   # Reward for Mutual Cooperation
    P = 5   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case1a():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 6   # Reward for Mutual Cooperation
    P = 2   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case2a():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 9   # Reward for Mutual Cooperation
    P = 8   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case3a():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 2   # Reward for Mutual Cooperation
    P = 1   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case1b():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 6   # Reward for Mutual Cooperation
    P = 5   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
def case2b():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 7   # Reward for Mutual Cooperation
    P = 6   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation

def case3b():
    global T, R, P, S
    T = 10  # Temptation to Defect
    R = 4   # Reward for Mutual Cooperation
    P = 3   # Punishment for Mutual Defection
    S = 10**-10   # Sucker's Payoff for Unilateral Cooperation
    
# Initial Population
population = {
    "homo_economicus": 1000,
    "homo_milliensis": 1000,
    "homo_kantiensis": 1000,
    "homo_smithiensis": 1000
}

# Set Population
def reset_population(E, M, K, S):
    population["homo_economicus"] = E
    population["homo_milliensis"] = M
    population["homo_kantiensis"] = K
    population["homo_smithiensis"] = S

# Individual Round
def simulate_round(population):
    total_population = sum(population.values())
    fitness = {key: 0 for key in population}
    for type1 in population:
        for type2 in population:
            # θ x θ Interactions
            interactions = population[type1] * population[type2]

            # Strategies
            
            strategy1 = 0
            strategy2 = 0
            u1 = 0

            ME1 = ((P / (R + P)) ** 2 * R) + ((P / (R + P)) * (R / (R + P)) * S) + ((P / (R + P)) * (R / (R + P)) * T) + ((R / (R + P)) ** 2 * P)
            ME2 = ((((2 * P - S - T) / (2 * P + 2 * R - 2 * S - 2 * T)) ** (2 * R)) + \
                             (((2 * P - S - T) / (2 * P + 2 * R - 2 * S - 2 * T)) * ((2 * R - S - T) / (2 * P + 2 * R - 2 * S - 2 * T)) * S) + \
                             (((2 * P - S - T) / (2 * P + 2 * R - 2 * S - 2 * T)) * ((2 * R - S - T) / (2 * P + 2 * R - 2 * S - 2 * T)) * T) + \
                             (((2 * R - S - T) / (2 * P + 2 * R - 2 * S - 2 * T)) ** (2 * P)))
            ME3 = ((2*P - S - T) / (2*P + 2*R - 2*S - 2*T)) * (P / (P + R)) * R + \
                            ((2*P - S - T) / (2*P + 2*R - 2*S - 2*T)) * (R / (P + R)) * S + \
                            ((2*R - S - T) / (2*P + 2*R - 2*S - 2*T)) * (P / (P + R)) * T + \
                            ((2*R - S - T) / (2*P + 2*R - 2*S - 2*T)) * (R / (P + R)) * P
            
            if 2 * R > S + T > 2 * P: #Case 1
                if type1 == "homo_economicus":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 1
                        strategy2 += 0
                elif type1 == "homo_milliensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0
                        strategy2 += 0
                elif type1 == "homo_kantiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 1
                        strategy2 += 0
                if type1 == "homo_smithiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME1
            elif 2 * R > 2 * P > S + T: #Case 2
                if type1 == "homo_economicus":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 1
                        strategy2 += 1
                elif type1 == "homo_milliensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME2
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME3
                elif type1 == "homo_kantiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0
                        strategy2 += 0
                if type1 == "homo_smithiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME3                        
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME1
            elif S + T > 2 * R > 2 * P: #Case 3
                if type1 == "homo_economicus":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 1
                        strategy2 += 1
                elif type1 == "homo_milliensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3)*S+(1/3)*T+(1/3)*ME2
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += ME3
                elif type1 == "homo_kantiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0
                        strategy2 += 0
                if type1 == "homo_smithiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += ME3
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME1
            elif 2 * R == S + T and S + T > 2 * P: #Case 4
                if type1 == "homo_economicus":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 1
                        strategy2 += 1
                elif type1 == "homo_milliensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * T
                    if type2 == "homo_kantiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * R + (1/2) * T
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * R + (1/2) * ME3
                elif type1 == "homo_kantiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * R + (1/2) * T
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0
                        strategy2 += 0
                if type1 == "homo_smithiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/2) * R + (1/2) * ME3
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME1
            elif 2 * R > S + T and S + T == 2 * P: #Case 5
                if type1 == "homo_economicus":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += 0.5 * S + 0.5 * P
                    if type2 == "homo_kantiensis":
                        strategy1 += 1
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 1
                        strategy2 += 1
                elif type1 == "homo_milliensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += 0.5 * S + 0.5 * P
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += 0.5 * R + 0.5 * P
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME3
                elif type1 == "homo_kantiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 0
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0
                        strategy2 += 0
                if type1 == "homo_smithiensis":
                    if type2 == "homo_economicus":
                        strategy1 += 1
                        strategy2 += 1
                    if type2 == "homo_milliensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME3
                    if type2 == "homo_kantiensis":
                        strategy1 += 0
                        strategy2 += 0
                    if type2 == "homo_smithiensis":
                        strategy1 += 0.5
                        strategy2 += 0.5
                        u1 += (1/3) * R + (1/3) * P + (1/3) * ME1
                
            # Payoffs
            u1 += R if strategy1 == 0 and strategy2 == 0 else 0
            u1 += S if strategy1 == 0 and strategy2 == 1 else 0 
            u1 += T if strategy1 == 1 and strategy2 == 0 else 0
            u1 += P if strategy1 == 1 and strategy2 == 1 else 0
            
            # Update Fitness
            fitness[type1] += u1 * interactions

    # Update Population
    total_fitness = sum(fitness.values())
    for type in population:
        population[type] = int((fitness[type] / total_fitness) * total_population)
    return population

# Iteration
def simulate(A):
    global population
    num_generations = A
    for _ in range(num_generations):
        population = simulate_round(population)
    return population

def test():
    reset_population(1000,1000,0,0)
    simulate(1000)
    print("e x m = ", population)

    reset_population(1000,0,1000,0)
    simulate(1000)
    print("e x k = ", population)

    reset_population(1000,0,0,1000)
    simulate(1000)
    print("e x s = ", population)

    reset_population(0,1000,1000,0)
    simulate(1000)
    print("m x k = ", population)

    reset_population(0,1000,0,1000)
    simulate(1000)
    print("m x s = ", population)

    reset_population(0,0,1000,1000)
    simulate(1000)
    print("k x s = ", population)

    reset_population(1000,1000,1000,0)
    simulate(1000)
    print("e x m x k = ", population)

    reset_population(1000,1000,0,1000)
    simulate(1000)
    print("e x m x s = ", population)

    reset_population(1000,0,1000,1000)
    simulate(1000)
    print("e x k x s = ", population)

    reset_population(1000,1000,1000,0)
    simulate(1000)
    print("m x k x s = ", population)

    reset_population(1000,1000,1000,1000)
    simulate(1000)
    print("e x m x k x s", population)

def edge_test(a,b):
    reset_population(a,b,0,0)
    simulate(1000)
    print("e x m = ", population)

    reset_population(a,0,b,0)
    simulate(1000)
    print("e x k = ", population)

    reset_population(a,0,0,b)
    simulate(1000)
    print("e x s = ", population)

    reset_population(b,a,0,0)
    simulate(1000)
    print("m x e = ", population)
    
    reset_population(0,a,b,0)
    simulate(1000)
    print("m x k = ", population)

    reset_population(0,a,0,b)
    simulate(1000)
    print("m x s = ", population)
    
    reset_population(b,0,a,0)
    simulate(1000)
    print("k x e = ", population)
    
    reset_population(0,b,a,0)
    simulate(1000)
    print("k x m = ", population)

    reset_population(0,0,a,b)
    simulate(1000)
    print("k x s = ", population)

    reset_population(b,0,0,a)
    simulate(1000)
    print("s x e = ", population)
    
    reset_population(0,b,0,a)
    simulate(1000)
    print("s x m = ", population)

    reset_population(0,0,b,a)
    simulate(1000)
    print("s x k = ", population)
    
def test_all():
    case1()
    print("Case 1: (10, 6, 4, 0)")
    test()

    case1a()
    print("Case 1a: (10, 6, 2, 0)")
    test()

    case1b()
    print("Case 1b: (10, 6, 5, 0)")
    test()
    
    case2()
    print("Case 2: (10, 8, 6, 0)")
    test()

    case2a()
    print("Case 2a: (10, 9, 8, 0)")
    test()

    case2b()
    print("Case 2b: (10, 7, 6, 0)")
    test()
    
    case3()
    print("Case 3: (10, 4, 2, 0)")
    test()
    
    case3a()
    print("Case 3a: (10, 2, 1, 0)")
    test()
    
    case3b()
    print("Case 3b: (10, 4, 3, 0)")
    test()
    
    case4()
    print("Case 4: (10, 5, 3, 0)")
    test()
    
    case5()
    print("Case 5: (10, 7, 5, 0)")
    test()
    
def test_skew(a,b):
    case1()
    print("Case 1: (10, 6, 4, 0)")
    edge_test(a,b)

    case1a()
    print("Case 1a: (10, 6, 2, 0)")
    edge_test(a,b)

    case1b()
    print("Case 1b: (10, 6, 5, 0)")
    edge_test(a,b)
    
    case2()
    print("Case 2: (10, 8, 6, 0)")
    edge_test(a,b)

    case2a()
    print("Case 2a: (10, 9, 8, 0)")
    edge_test(a,b)

    case2b()
    print("Case 2b: (10, 7, 6, 0)")
    edge_test(a,b)
    
    case3()
    print("Case 3: (10, 4, 2, 0)")
    edge_test(a,b)
    
    case3a()
    print("Case 3a: (10, 2, 1, 0)")
    edge_test(a,b)
    
    case3b()
    print("Case 3b: (10, 4, 3, 0)")
    edge_test(a,b)
    
    case4()
    print("Case 4: (10, 5, 3, 0)")
    edge_test(a,b)
    
    case5()
    print("Case 5: (10, 7, 5, 0)")
    edge_test(a,b)

test_all()
```
